## 1. Load HuggingFace Dataset trực tiếp Stream
Không cần tải 100GB về máy, dùng Streaming để stream từng bài hát hoặc phổ âm thẳng qua Ram!


```python
dataset_repo = 'Khahn-nh/fma-small-1'
# Dùng streaming=True thì model có thể get item liên tục
ds_stream = load_dataset(dataset_repo, split='train', streaming=True)

iterator = iter(ds_stream)
first_item = next(iterator)

print("Keys available:", first_item.keys())
print("Label:", first_item['label'])
```
