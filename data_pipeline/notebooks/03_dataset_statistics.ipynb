# Phân tích Thống kê & Chuẩn hóa (Normalization)
Notebook này hình dung việc tính toán giá trị Mean/Std tập trung toàn cục (global dataset), qua đó giúp Model huấn luyện tốt hơn nhờ Z-score normalization.


```python
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os

# Xác định môi trường
IN_COLAB = 'google.colab' in sys.modules
if IN_COLAB:
    SRC_DIR = '/content/music-data/data_pipeline'
else:
    from pathlib import Path
    SRC_DIR = str(Path().resolve().parent)
sns.set_theme(style="whitegrid")
```

## Tạo Dữ liệu Giả lập (Nếu file JSON chưa được tạo)
Ta dùng data random để demo trực quan khái niệm này.


```python
stats_mean = np.random.normal(-50, 15, size=128)
stats_std = np.random.uniform(5, 20, size=128)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].plot(stats_mean, color='royalblue')
axes[0].set_title('Global Mean per Mel Bin')
axes[0].set_xlabel('Mel Bin Index')
axes[0].set_ylabel('Mean value')
axes[1].plot(stats_std, color='darkorange')
axes[1].set_title('Global Standard Deviation per Mel Bin')
axes[1].set_xlabel('Mel Bin Index')
axes[1].set_ylabel('Std value')
plt.tight_layout()
plt.show()
```

## Minh họa việc Chuẩn hóa (Z-Score)


```python
raw_sample = np.random.normal(stats_mean, stats_std)
normalized_sample = (raw_sample - stats_mean) / (stats_std + 1e-6)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(raw_sample, ax=ax1, bins=30, kde=True)
ax1.set_title('Phân phối gốc (trước khi Normalize)')
sns.histplot(normalized_sample, ax=ax2, bins=30, kde=True, color='green')
ax2.set_title('Phân phối Z-score (Mean ~ 0, Std ~ 1)')
plt.tight_layout()
plt.show()
```
